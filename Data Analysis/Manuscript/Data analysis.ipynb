```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
df.info()
```

```python
df['HomeTeam'].value_counts()
```

```python
# Convert the 'date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year, month, and weekday
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['weekday'] = df['Date'].dt.day_name()  # This will give the weekday as string

df.head(20)
```

```python
# Convert 'date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Filter the DataFrame for rows starting from '2024-07-26'
filtered_df = df[df['Date'] >= '2024-07-26']

# Count occurrences of each team in 'HomeTeam' from the filtered data
team_counts = filtered_df['HomeTeam'].value_counts()

# Display the result
print("Number of occurrences for each team from 2024-07-26:")
print(team_counts)
```

```python


# Define the date range
start_date = '2023-07-28'
end_date = '2024-07-26'

# Filter the DataFrame for the date range (excluding 2024-07-26)
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] < end_date)]

# Count how many unique teams are in 'HomeTeam' within this date range
team_count = filtered_df['HomeTeam'].nunique()

# Display the result
print(f"Number of unique teams from {start_date} to {end_date} (excluding {end_date}): {team_count}")
```

```python

# Group by 'HomeTeam' and get the list of unique 'year' values for each team
team_years = df.groupby('HomeTeam')['year'].apply(lambda x: list(x.unique()))
# Display the result
print("Unique 'year' values for each team:")
print(team_years)
```

```python
# Convert 'time' column to datetime
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time

# Function to categorize the time
def categorize_time(t):
    if t >= pd.to_datetime('12:00:00').time() and t < pd.to_datetime('13:30:00').time():
        return 'noon'
    elif t >= pd.to_datetime('13:30:00').time() and t < pd.to_datetime('17:00:00').time():
        return 'afternoon'
    elif t >= pd.to_datetime('17:00:00').time() and t < pd.to_datetime('19:30:00').time():
        return 'evening'
    elif t >= pd.to_datetime('19:30:00').time() or t < pd.to_datetime('06:00:00').time():
        return 'night'
    else:
        return 'morning'

# Apply the function to the 'time' column
df['time_of_day'] = df['Time'].apply(categorize_time)
df.head()
```

```python
#make a data frame contains the features of matches 
selected_columns = ['year', 'month', 'weekday', 'time_of_day', 'HomeTeam', 'AwayTeam', 'HTHG', 'HTAG', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR', 'FTR'] 
df_match = df[selected_columns]
df_match = df_match.dropna()
```

```python
#Import necessary libraries
import pandas as pd
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
from sklearn.preprocessing import LabelEncoder

#make a data frame contains the features of matches 
selected_columns = ['year', 'month', 'weekday', 'time_of_day', 'HomeTeam', 'AwayTeam', 'HTHG', 'HTAG', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR', 'FTR'] 
df_match = df[selected_columns]
df_match = df_match.dropna()

X = df_match.drop('FTR', axis=1)
y = df_match['FTR'] 

#Encode target and categorical features (if not already encoded)
labelencoder = LabelEncoder()
X['weekday'] = labelencoder.fit_transform(X['weekday'])
X['time_of_day'] = labelencoder.fit_transform(X['time_of_day'])
X['HomeTeam'] = labelencoder.fit_transform(X['HomeTeam'])
X['AwayTeam'] = labelencoder.fit_transform(X['AwayTeam'])



y_encoded = labelencoder.fit_transform(y)

# Apply Chi-Squared test
chi2_selector = SelectKBest(chi2, k='all')  
X_kbest = chi2_selector.fit(X, y_encoded)

#  Get the scores and p-values
chi2_scores = X_kbest.scores_
chi2_pvalues = X_kbest.pvalues_

# Create a DataFrame to summarize the results
feature_scores = pd.DataFrame({
    'Feature': X.columns,
    'Chi2 Score': chi2_scores,
    'P-value': chi2_pvalues
})

# Function to filter features based on Chi2 and p-value thresholds
def filter_features(df, chi2_threshold, pvalue_threshold):
    filtered_df = df[(df['Chi2 Score'] > chi2_threshold) & (df['P-value'] < pvalue_threshold)]
    return filtered_df

# Apply the filter with your thresholds
chi2_threshold = 10
pvalue_threshold = 0.05
filtered_features = filter_features(feature_scores, chi2_threshold, pvalue_threshold)

# Display the filtered features
print(filtered_features)

```

```python
selected_columns = ['HomeTeam', 'AwayTeam', 'HTHG', 'HTAG', 'HS', 'AS', 'HST', 'AST', 'HY', 'HR', 'AR', 'FTR'] 
df_ML = df[selected_columns]
df_ML = df_ML.dropna()

X_ML= df_ML.drop('FTR', axis=1)
y_ML= df_ML['FTR'] 

#Encode target and categorical features (if not already encoded)
labelencoder = LabelEncoder()
X_ML['HomeTeam'] = labelencoder.fit_transform(X_ML['HomeTeam'])
X_ML['AwayTeam'] = labelencoder.fit_transform(X_ML['AwayTeam'])


print(X_ML)
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = sm.add_constant(X_ML)

# Calculate VIF for each variable
vif = pd.DataFrame()
vif['Variable'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Encode categorical variables if needed
X_ML = pd.get_dummies(X_ML)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_ML, y_ML, test_size=0.30, random_state=42)

# Initialize and train the mode
model = RandomForestClassifier(n_estimators=300, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize the logistic regression model
model = LogisticRegression(penalty='l2', multi_class='multinomial', solver='lbfgs', C=0.2, max_iter=1000, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Print the classification report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l2'],  # For multinomial regression, L1 penalty is not supported with lbfgs solver
    'solver': ['lbfgs']
}

# Initialize GridSearchCV
grid_search = GridSearchCV(LogisticRegression(multi_class='multinomial', max_iter=1000), param_grid, cv=5, verbose=1)
grid_search.fit(X_train, y_train)

# Print best parameters
print("Best Parameters:", grid_search.best_params_)

# Use the best estimator
best_model = grid_search.best_estimator_

# Evaluate the model
y_pred = best_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Use class weights to handle class imbalance
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', class_weight='balanced', max_iter=1000)
model.fit(X_train, y_train)

# Evaluate the model
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
